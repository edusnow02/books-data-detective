```python
import pandas as pd

tables = pd.DataFrame([
    ["categories", "id", "INTEGER", "PK"],
    ["categories", "name", "TEXT", "UNIQUE"],

    ["books", "id", "INTEGER", "PK"],
    ["books", "title", "TEXT", ""],
    ["books", "price", "REAL", ""],
    ["books", "rating", "INTEGER", ""],
    ["books", "description", "TEXT", ""],
    ["books", "category_id", "INTEGER", "FK -> categories.id"],
    ["books", "url", "TEXT", "UNIQUE"],

    ["authors", "id", "INTEGER", "PK"],
    ["authors", "name", "TEXT", "UNIQUE"],

    ["books_authors", "book_id", "INTEGER", "FK -> books.id"],
    ["books_authors", "author_id", "INTEGER", "FK -> authors.id"],
], columns=["Table", "Column", "Type", "Constraint"])

relations = pd.DataFrame([
    ["books.category_id", "categories.id", "Many to One"],
    ["books_authors.book_id", "books.id", "Many to Many"],
    ["books_authors.author_id", "authors.id", "Many to Many"]
], columns=["From", "To", "Relationship"])

print("📦 TABLE STRUCTURE")
display(tables)

print("\n🔗 RELATIONSHIPS")
display(relations)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.axis('off')

text = """
CATEGORIES
- id (PK)
- name

BOOKS
- id (PK)
- title
- price
- rating
- description
- url
- category_id (FK)

AUTHORS
- id (PK)
- name

BOOKS_AUTHORS
- book_id (FK)
- author_id (FK)
"""

ax.text(0.1, 0.5, text, fontsize=12)
plt.show()
```
